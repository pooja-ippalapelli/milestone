```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('ZHVI_dataset.csv')
newyork_set = df[df['City'] == 'New York']
newyork_set.head()
```

```python
# Selecting the columns containing housing price data
price_columns = df.columns[9:]  # Starting from the 10th column onwards

# Scatter plot to compare multiple neighborhoods on a specific date (e.g., 12/31/2023)
date_column = '12/31/2023'
scatter_data = df[['RegionName', date_column]]

fig = px.scatter(scatter_data, x='RegionName', y=date_column, title=f'Housing Prices on {date_column}')
fig.update_xaxes(title_text='Neighborhood')
fig.update_yaxes(title_text='Housing Price')
fig.show()
```

```python
# Calculating the average housing prices over time
average_prices_over_time = df[price_columns].mean()

# Creating a line plot to visualize the average housing prices over time
fig = px.line(x=price_columns, y=average_prices_over_time, title='Average Housing Prices Over Time')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Average Housing Price')
fig.show()
```

```python
# Grouping the data by neighborhood and calculating the average prices
average_prices_by_neighborhood = df.groupby('RegionName')[price_columns].mean().reset_index()

# Reshaping the data to have neighborhoods as rows and dates as columns
average_prices_melted = average_prices_by_neighborhood.melt(id_vars=['RegionName'], var_name='Date', value_name='Average Price')

# Creating a line plot to visualize the average housing prices over time, grouped by neighborhood
fig = px.line(average_prices_melted, x='Date', y='Average Price', color='RegionName', title='Average Housing Prices Over Time by Neighborhood')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Average Housing Price')
fig.show()
```
